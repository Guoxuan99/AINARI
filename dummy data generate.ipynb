```python
data_dict = collections.defaultdict(dict)
```

```python
size = 100
risks = ['Brown Spot','Leaf Blast','Hispa','Flood','Drought','Healthy']

for risk in risks:
  if risk == "Brown Spot":
    temp = np.round_(np.random.uniform(16, 36, size),decimals = 2) #temp 16-36
    hum = np.round_(np.random.uniform(86, 100, size),decimals = 2) #hump 86-100
    wl = np.round_(np.random.uniform(5, 50, size),decimals = 2) #water level: low(around 5)
    n = np.round_(np.random.uniform(40, 55, size),decimals = 2) #N: low(60-10)
    p = np.round_(np.random.uniform(15, 30, size),decimals = 2) #P: low(35-10)
    k = np.round_(np.random.uniform(15, 30, size),decimals = 2) #K: low(35-10)
    ph = np.round_(np.random.uniform(2, 4, size),decimals = 2) #ph 2-4
    rf = np.round_(np.random.uniform(150, 172, size),decimals = 2) #rf: low(182-10)

  elif risk == "Leaf Blast":
    temp = np.round_(np.random.uniform(25, 28, size),decimals = 2) #temp 25-28
    hum = np.round_(np.random.uniform(90, 100, size),decimals = 2) #hump 90-100
    wl = np.round_(np.random.uniform(525, 560, size),decimals = 2) #water level: high(around 530)
    n = np.round_(np.random.uniform(105, 120, size),decimals = 2) #N: high(99+10)
    p = np.round_(np.random.uniform(70, 90, size),decimals = 2) #P: high(60+10)
    k = np.round_(np.random.uniform(55, 80, size),decimals = 2) #K: high(45+10)
    ph = np.round_(np.random.uniform(5.1, 5.9, size),decimals = 2) #ph 5.1-5.9
    rf = np.round_(np.random.uniform(310, 340, size),decimals = 2) #rf: high(300+10)

  elif risk == "Hispa":
    temp = np.round_(np.random.uniform(16, 36, size),decimals = 2) #temp 16-36
    hum = np.round_(np.random.uniform(85, 100, size),decimals = 2) #hump 85-100
    wl = np.round_(np.random.uniform(525, 560, size),decimals = 2) #water level: high(around 530)
    n = np.round_(np.random.uniform(105, 120, size),decimals = 2) #N: high(99+10)
    p = np.round_(np.random.uniform(70, 90, size),decimals = 2) #P: high(60+10)
    k = np.round_(np.random.uniform(55, 80, size),decimals = 2) #K: high(45+10)
    ph = np.round_(np.random.uniform(6.4, 7, size),decimals = 2) #ph around 6.8
    rf = np.round_(np.random.uniform(310, 340, size),decimals = 2) #rf: high(300+10)
  
  elif risk == "Flood":
    temp = np.round_(np.random.uniform(16, 28, size),decimals = 2) #temp 16-28
    hum = np.round_(np.random.uniform(86, 100, size),decimals = 2) #hump 86-100
    wl = np.round_(np.random.uniform(525, 560, size),decimals = 2) #water level: high(around 530)
    n = np.round_(np.random.uniform(70, 90, size),decimals = 2) #N: normal(80)
    p = np.round_(np.random.uniform(35, 55, size),decimals = 2) #P: normal(47)
    k = np.round_(np.random.uniform(30, 50, size),decimals = 2) #K: normal(40)
    ph = np.round_(np.random.uniform(6.3, 6.9, size),decimals = 2) #ph 6.4-6.8
    rf = np.round_(np.random.uniform(310, 340, size),decimals = 2) #rf: high(300+10)
  
  elif risk == "Drought":
    temp = np.round_(np.random.uniform(25, 36, size),decimals = 2) #temp 25-36
    hum = np.round_(np.random.uniform(50, 60, size),decimals = 2) #hump (80-20)
    wl = np.round_(np.random.uniform(0, 0.1, size),decimals = 2) #water level: low(around 0)
    n = np.round_(np.random.uniform(70, 90, size),decimals = 2) #N: normal(80)
    p = np.round_(np.random.uniform(35, 55, size),decimals = 2) #P: normal(47)
    k = np.round_(np.random.uniform(30, 50, size),decimals = 2) #K: normal(40)
    ph = np.round_(np.random.uniform(3, 10, size),decimals = 2) #ph 3-10
    rf = np.round_(np.random.uniform(50, 100, size),decimals = 2) #rf: low(182-30)
  
  elif risk == "Healthy":
    temp = np.round_(np.random.uniform(20, 28, size),decimals = 2) #temp 20-28
    hum = np.round_(np.random.uniform(80, 85, size),decimals = 2) #hump (80-85)
    wl = np.round_(np.random.uniform(350, 450, size),decimals = 2) #water level: low(around 0)
    n = np.round_(np.random.uniform(70, 90, size),decimals = 2) #N: normal(80)
    p = np.round_(np.random.uniform(35, 55, size),decimals = 2) #P: normal(47)
    k = np.round_(np.random.uniform(30, 50, size),decimals = 2) #K: normal(40)
    ph = np.round_(np.random.uniform(5, 8, size),decimals = 2) #ph 3-10
    rf = np.round_(np.random.uniform(180, 300, size),decimals = 2) #rf: low(182-30)

  else:
    pass
  
  data_dict[risk]['temp'] = temp.tolist()
  data_dict[risk]['hum'] = hum.tolist()
  data_dict[risk]['wl'] = wl.tolist()
  data_dict[risk]['n'] = n.tolist()
  data_dict[risk]['p'] = p.tolist()
  data_dict[risk]['k'] = k.tolist()
  data_dict[risk]['ph'] = ph.tolist()
  data_dict[risk]['rf'] = rf.tolist()
```

```python
dataList = []
for key, val in data_dict.items():
  
  risk = key
  val_temp = val['temp']
  val_hum = val['hum']
  val_wl = val['wl']
  val_n = val['n']
  val_p = val['p']
  val_k = val['k']
  val_ph = val['ph']
  val_rf = val['rf']

  for temp, hum, wl, n, p, k, ph, rf in zip(val_temp, val_hum, val_wl, val_n, val_p, val_k, val_ph, val_rf):
    dataList.append([risk, temp, hum, wl, n, p, k, ph, rf])
  
data_df = pd.DataFrame(dataList, columns = ['Risk', 'Temperature', 'Humidity', 'Water Level', 'Nitrogen', 
                                            'Phosphorus', 'Potasium', 'pH', 'Rainfall'])
```
